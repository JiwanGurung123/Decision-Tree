### Importing the libraries

```python
import sklearn.datasets as dataset
import pandas as pd
```

### Importing the iris dataset

```python
iris=dataset.load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
X=df
y=iris.target
print(y)
```

### Splitting the dataset: training set and testing set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

### Decision Tree Model

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
print('Training complete!!')
```

### Comparing the actual and predicted results

```python
y_pred=model.predict(X_test)
df1=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
df1.head()
```

### Importing libraries to display the decision tree

```python
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
from io import StringIO
```

```python
dot_data = StringIO()
export_graphviz(model, out_file=dot_data, feature_names=iris.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
```

```python
Image(graph.create_jpg())
```

### Testing the model with a sample and predicting the class of the sample

```python
#Iris-setosa-[0]
#Iris-versicolor-[1]
#Iris-virginica-[2]

sample=[4,5,3,2]
result=model.predict([sample])
print(result)
```
